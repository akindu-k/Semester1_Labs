Exercise 01
Develop a program to input three angles of a triangle and output the type of the triangle. First, you have to check whether the given angles form a triangle or not. If they do form a triangle then output the type of the triangle as "Right angled", "Obtuse angled" or "Acute angled" triangle. If the angles do not form a triangle you can output “Angles do not form a triangle”.

```python
#getting user inputs
a = int(input("Enter angle 1: "))
b = int(input("Enter angle 2: "))
c = int(input("Enter angle 3: "))

if (a+b+c)!= 180 or (a==0 or b==0 or c==0): #checking whether a,b,c values can form a triangle
    print("Angles do not form a triangle")
elif (a==90 or b==90 or c==90): #sufficient condtion for right angled triangles
    print("Right angled")
elif (a>90 or b>90 or c>90): #sufficient condtion for obtuse angled triangles
    print("Obtuse angled")
else:
    print("Acute angled")
```
